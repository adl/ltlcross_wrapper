```python
import pandas as pd

from ltlcross_wrapper import ResAnalyzer, remove_tools, merge_resfiles, update_resfile
```

```python
names = [f"{s}_{t}" for s in ["literature","random"] for t in ["det","sd","nd"]]
```

```python
benchmarks = {}
for name in names:
    b = ResAnalyzer(f"test_merging/{name}.csv",  cols=["states","time","acc","transitions","edges"])
    b.name = name
    b.orig_count = len(b.values)
    b.clean_count = len(b.values.dropna())
    benchmarks[name] = b
```

```python
new_data = {}
for name in names:
    b = ResAnalyzer(f"test_merging/{name}.no.ncsb.csv",  cols=["states","time","acc","transitions","edges"])
    b.name = name
    b.orig_count = len(b.values)
    b.clean_count = len(b.values.dropna())
    new_data[name] = b
```

```python
b = benchmarks["literature_nd"]
b.cumulative()
```

```python
n = new_data["literature_nd"]
n.cumulative()
```

```python
tools = n.tools
orig_f = "test_merging/literature_nd.csv"
add_f = "test_merging/literature_nd.no.ncsb.csv"
out_f = "test_merging/literature_nd.new.csv"
fin_f = "test_merging/literature_nd.fin.csv"
```

```python
remove_tools(orig_f, out_f, tools)
ResAnalyzer(out_f).cumulative()
```

```python
update_resfile(orig_f, add_f, fin_f)
ResAnalyzer(fin_f).cross_compare()
```
